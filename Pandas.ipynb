# Ejercicio 1:
Construir un objeto de series de tiempo con el índice igual al año de los juegos olimpicos.

```python
# Cargo DataFrame
datos_crudos = pd.read_csv('Summer-Olympic-medals-1976-to-2008.csv')
# Copio el original para poder probar transformaciónes
Olimpiadas = datos_crudos.copy()
Olimpiadas.tail()
```

```python
shape_antes = Olimpiadas.shape
```

```python
# Remuevo todos los valores nan y reseteo el index, para poder hacer 
# loops mas adelante sin que salte un error:
Olimpiadas = Olimpiadas.dropna()
Olimpiadas.reset_index(drop=True, inplace=True)
shape_despues = Olimpiadas.shape
```

```python
# Viendo la diferencia, vemos que hay 117 "nan" que fueron eliminados:
shape_antes[0] - shape_despues[0]
```

```python
# Veo que tipo de datos tiene la columna "Year" y tranformo de float64 a 
# int64:
Olimpiadas['Year'] = Olimpiadas['Year'].astype('int64')
Olimpiadas
```

```python
# Defino la serie que va a contener las fechas de cada participación en 
# las olimpiadas, transformada a formato Datetime y filtrando solamente 
# el año:
fechas= pd.DatetimeIndex(Olimpiadas['Year']).year
fechas
```

```python
# Creo la serie de pandas, que utiliza como indice la fecha de participacion de cada atleta:
Olimpiadas_participacion = pd.Series(Olimpiadas['Athlete'].values, index = fechas)
Olimpiadas_participacion
```

# Ejercicio 2:
Separar nombres de apellidos en dos columnas distintas usando la coma como separador.

```python
# Defino la serie que va a contener la columna "Athlete" del dataframe
NomAp = Olimpiadas['Athlete']
NomAp.head()
```

```python
# Hago el split entre nombre y apellido. Todavia no voy a crear las 
# variables con nombre y apellido. Lo hago para obtener 2 objetos en el 
# caso de tener nombre y apellido, y 1 cuando no tengo nombre o apellido.
# Hago esto para mas adelante poder eliminar estos casos, ya que mas 
# adelante generaran error al intentar agregar las nuevas columnas al 
# dataframe:
NomAp_separado = NomAp.str.split(',')
NomAp_separado.reset_index()
NomAp_separado
```

```python
# Hago un loop, que cada vez que acceda a un indice la serie NomAp_separado
# que sea diferente de 2, elimine esa fila del dataframe:
for h in range(len(NomAp_separado)):
    if len(NomAp_separado[h]) != 2:
         Olimpiadas.drop([h], inplace = True)
```

```python
# Vuelvo a resetear el indice para no tener problemas con los loos que se 
# hacen mas adelante:
Olimpiadas.reset_index(drop=True, inplace=True)
```

```python
# Nuevamente, redefino la serie que va a contener apellidos y nombres de 
# atletas, pero esta vez sin las filas que tienen algun valor faltante.
# Tambien hago el split nuevamente entre nombres y apellidos, ocupando
# las mismas variables que antes.
NomAp = Olimpiadas['Athlete']
NomAp.head()
NomAp_separado = NomAp.str.split(',')
NomAp_separado.reset_index()
NomAp_separado
```

```python
# Genero la lista con los valores de todos los apellidos en la serie NamAp:
Apellidos = []
for i in range(len(NomAp_separado)):
    Apellidos.append(NomAp_separado[i][0])
    i += 1
Apellidos
```

```python
# Genero la lista con los valores de todos los nombres en la serie NamAp:
Nombres = []
for j in range(len(NomAp_separado)):
    Nombres.append(NomAp_separado[j][1])
    j += 1
Nombres
```

```python
# Confirmo que tengan el mismo tamaño:
print(len(Apellidos))
print(len(Nombres))
```

```python
# Finalmente, introduzco las columnas separadas de Nombres y Apellidos en 
# el dataframe. Tambien elimino la columna 'Athlete' Ya que no va a hacer
# falta:
Olimpiadas_final = Olimpiadas.copy()
Olimpiadas_final['Nombres'] = Nombres
Olimpiadas_final['Apellidos'] = Apellidos
Olimpiadas_final.drop(['Athlete'], axis = 1)
Olimpiadas_final.head()
```
