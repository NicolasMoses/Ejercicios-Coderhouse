# Ejercicio 2:
Realizar un histograma con los salarios. Qué rango(s) de salarios son los más populares?

```python
plt.hist(Datos['Salary'], bins=20)
plt.xlabel('Salarios', size = 15);
```

Los salarios mas comunes estan entre 50000 y 11000. Particularmente, alrededor de 60000 es donde hay mayor acumulación de valores

# Ejercicio 3:
Realizar un gráfico de violín con los salarios, discriminados por género. 

```python
sns.catplot(data=Datos, kind="violin", x="GenderID", y="Salary", split=True);

# Se nota la presencia de outliers en los valores mas altos para ambos generos, pero mas para el 0.
```
